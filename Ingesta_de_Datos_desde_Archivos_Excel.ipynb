```python
# Importamos librerias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importacion de un excel!
# leemos el archivo de excel y lo guardamos el contenido del archivo excel en un DataFrame llamado data_excel

data_excel = pd.read_excel("/content/data.xlsx")

# mostramos el contenido del DataFrame

data_excel
```

```python
# Ahora vamos a explorar y analizar los datos cargados desde el archivo de Excel

# Ver las primeras filas del dataframe
print("Primeras filas del dataframe:")
print(data_excel.head())
print()
# Información general del dataframe
print("\nInformación general sobre el dataframe:")
print(data_excel.info())
print()
# Estadísticas descriptivas
print("\nEstadísticas descriptivas:")
print(data_excel.describe())

# Comprobar valores nulos
print("\nValores nulos por columna:")
print(data_excel.isnull().sum())
```

```python
# Vamos a analizar y explorar los datos de forma visual, utilizando gráficos para entender mejor la distribución, correlaciones y posibles valores atípicos en el DataFrame.

# Ver distribución de las variables numéricas
print("\nDistribución de las variables numéricas:")
data_excel.hist(bins=20, figsize=(14,10))
plt.suptitle("Distribución de Variables Numéricas")
plt.show()

# Correlación entre variables numéricas usando numeric_only
print("\nMatriz de correlación:")
corr = data_excel.corr(numeric_only=True)  # O usar select_dtypes para filtrar columnas numéricas
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title("Mapa de Calor de Correlación")
plt.show()

# Ver la cantidad de valores únicos por columna
print("\nValores únicos por columna:")
print(data_excel.nunique())

# Graficar la distribución de una variable específica
# (por ejemplo, si hay una columna llamada 'edad', para ver la distribución de la edad)
if 'edad' in data_excel.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data_excel['edad'], kde=True, color='blue')
    plt.title('Distribución de la Edad')
    plt.xlabel('Edad')
    plt.ylabel('Frecuencia')
    plt.show()

# Verificación de outliers (valores atípicos) usando boxplot
# Si existe una columna llamada 'ingresos', para ver los outliers en ella
if 'ingresos' in data_excel.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=data_excel['ingresos'])
    plt.title('Outliers en Ingresos')
    plt.show()
```

# Comparativa entre CSV y Excel

Los formatos **CSV** (Comma-Separated Values) y **Excel** son ampliamente utilizados para almacenar y gestionar datos, pero presentan diferencias significativas en términos de funcionalidad, características y usos. A continuación, se realiza una comparación detallada de ambos:

## 1. Estructura y Formato
- **CSV**:
  - Es un archivo de texto plano donde los valores están separados por comas (u otros delimitadores como tabulaciones o punto y coma).
  - Cada línea representa un registro, y cada campo dentro del registro está separado por un delimitador.
  - No almacena formatos, fórmulas, ni imágenes.
  
- **Excel**:
  - Es un formato propietario de Microsoft (XLS, XLSX) que permite almacenar datos en hojas de cálculo con filas y columnas.
  - Soporta fórmulas, formatos de celda (colores, estilos de texto, etc.), gráficos, tablas dinámicas y otros elementos avanzados.
  - Es más adecuado para trabajos complejos, con múltiples hojas y funcionalidades interactivas.

## 2. Facilidad de Uso
- **CSV**:
  - Es fácil de leer y escribir debido a su simplicidad.
  - Los archivos CSV pueden abrirse con cualquier editor de texto o programa de hojas de cálculo.
  
- **Excel**:
  - Requiere software especializado, como Microsoft Excel o Google Sheets, para acceder a su contenido.
  - Ofrece herramientas interactivas, como la creación de gráficos y análisis de datos, pero estas requieren un aprendizaje adicional.

## 3. Tamaño y Eficiencia
- **CSV**:
  - Los archivos CSV tienden a ser más ligeros, ya que no contienen datos adicionales como formatos o imágenes.
  - Son adecuados para datasets grandes cuando no es necesario mantener una estructura compleja.
  
- **Excel**:
  - Los archivos Excel pueden ser más grandes debido a los elementos adicionales que contienen (fórmulas, formatos, macros, etc.).
  - Sin embargo, Excel tiene límites de filas y columnas (aunque muy amplios), lo que puede ser un inconveniente para archivos extremadamente grandes.

## 4. Compatibilidad
- **CSV**:
  - El formato CSV es muy compatible entre diferentes sistemas operativos y aplicaciones (Windows, macOS, Linux, Google Sheets, Python, R, etc.).
  - Debido a su simplicidad, los archivos CSV son fácilmente importables y exportables desde diferentes lenguajes de programación y bases de datos.

- **Excel**:
  - Requiere software específico para abrir archivos XLSX, aunque hay alternativas como Google Sheets y otras aplicaciones de hojas de cálculo.
  - Si bien Excel es ampliamente usado, su compatibilidad con programas de código abierto o ciertos lenguajes de programación puede ser limitada en comparación con CSV.

## 5. Funcionalidades Avanzadas
- **CSV**:
  - No soporta funciones avanzadas como fórmulas, gráficos o macros. Solo almacena datos.
  
- **Excel**:
  - Permite el uso de funciones avanzadas, como:
    - **Fórmulas**: Cálculos automáticos y complejos.
    - **Tablas dinámicas**: Análisis avanzado de datos.
    - **Gráficos interactivos**: Visualización de datos.
    - **Macros**: Automatización de tareas repetitivas.

## 6. Usos Comunes
- **CSV**:
  - Ideal para el intercambio de datos entre sistemas diferentes o para almacenar grandes volúmenes de datos de manera eficiente.
  - Utilizado frecuentemente en análisis de datos, bases de datos y programación.

- **Excel**:
  - Ampliamente usado en tareas que requieren análisis visual, como la contabilidad, presupuestos, análisis financiero, reportes, etc.
  - Es una herramienta ideal para quienes necesitan funcionalidades avanzadas para manipular y analizar datos de manera interactiva.

## 7. Desventajas
- **CSV**:
  - No soporta formatos ni funciones avanzadas (fórmulas, gráficos, tablas dinámicas).
  - No es ideal para almacenar grandes cantidades de datos complejos o interrelacionados.

- **Excel**:
  - Puede ser más pesado y lento con grandes volúmenes de datos.
  - La dependencia de software propietario (Microsoft Excel) puede ser un inconveniente para algunos usuarios.
  
## Resumen Final
- El formato **CSV** es ideal cuando se necesita simplicidad, eficiencia en el tamaño del archivo, y compatibilidad entre diferentes sistemas y lenguajes.
- El formato **Excel** es más adecuado para tareas interactivas y análisis de datos complejos que requieren visualización avanzada, automatización y fórmulas.
